```python
def splitDatetime(df) :
    cols=df.select_dtypes(include = ['datetime64']).columns.values.tolist()
    for i in cols:
        df['year'] = df[i].apply(lambda l: l.year)
        df['month'] = df[i].apply(lambda l: l.month)
        df['day'] = df[i].apply(lambda l: l.day)
        df['hour'] = df[i].apply(lambda l: l.hour)
        #df[str(i)+'date'] = df[i].apply(lambda l: l.date())
    return df
train=splitDatetime(train)
```

```python
ax=train[train['Train_Flag']==1].set_index('InvoiceDate')['Quantity'].plot(c='g',figsize=(18,8))
train[train['Train_Flag']==0].set_index('InvoiceDate')['Quantity'].plot(c='r',ax=ax)
```
